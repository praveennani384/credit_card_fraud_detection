```python
import pandas as pd
import numpy as np
import joblib
import warnings
warnings.filterwarnings("ignore")

# Load the trained model
model = joblib.load('credit_card_fraud_detection_model.pkl')

# Define a function to make predictions
def predict(input_data):
    prediction = model.predict(input_data)
    return prediction

# Define function to print prediction result
def print_prediction(prediction):
    if prediction == 0:
        print('It\'s a legit Transaction')
    else:
        print('It\'s a Fraud Transaction')

# Generate sample input data and make predictions 50 times
for _ in range(100):
    data = {
        'Time': np.random.uniform(low=0, high=172800),
        **{f'V{i}': np.random.randn() for i in range(1, 29)},
        'Amount': np.random.uniform(low=0, high=1000)
    }
    input_data = pd.DataFrame(data, index=[0])  # Ensure it's a single-row DataFrame
    
    # Make predictions using the deployed model
    prediction = predict(input_data)
    print("Prediction:", prediction[0])
    print_prediction(prediction[0])
    print('-' * 50)

    # Combine input data with prediction
    result_data = pd.concat([input_data, pd.DataFrame([prediction], columns=['Class'])], axis=1)

    # Save the data to an Excel file
    if not isinstance(result_data, pd.DataFrame):
        result_data = pd.DataFrame(result_data)

    try:
        # Load existing predictions (if any)
        existing_predictions = pd.read_excel('predictions.xlsx')

        # Append new results to existing data
        updated_data = existing_predictions.append(result_data, ignore_index=True)

    except (FileNotFoundError, IOError) as e:
        # Create a new file if not found or there's an IO error
        updated_data = result_data

    # Save the updated data back to the Excel file
    updated_data.to_excel('predictions.xlsx', index=False)

    print("Data and prediction appended to predictions.xlsx")
```
